```python
# 릿지, 라쏘, SVM

import pandas as pd

df1 = pd.read_csv('../Data/All.csv')
df2 = pd.read_csv('../Data/burgerking.csv')
df3 = pd.read_csv('../Data/lotteria.csv')
df4 = pd.read_csv('../Data/Macdonald.csv')
df5 = pd.read_csv('../Data/NoBrand.csv')
df6 = pd.read_csv('../Data/prank.csv')
```

```python
common_colname=["restaurant", "item","calories","sodium","sugar","total_fat","protein"]
df1=df1[common_colname]
```

```python
df2["restaurant"]="burgerking"
# 이상한 공백 값이 있어서, 직접 조회하여 열 순서 바꿈
new_df2=df2[[df2.columns[-1],df2.columns[0],df2.columns[1],df2.columns[3],df2.columns[4],df2.columns[5],df2.columns[2]]]
new_df2.columns=common_colname
```

```python
# sodium, total_fat, protein에 있는 괄호 값 제거하기
def delete_paran(df) :
    for i in range(df["sodium"].shape[0]):
        for string in ("sodium","total_fat","protein"):
            data=df[string][i]
            index=data.find("(")
            df[string][i]=data[:index]
delete_paran(new_df2)
```

```python
df3["restaurant"]="lotteria"
# 이상한 공백 값이 있어서, 직접 조회하여 열 순서 바꿈
new_df3=df3[[df3.columns[-1],df3.columns[0],df3.columns[3],df3.columns[5],df3.columns[6],df3.columns[7],df3.columns[4]]]
new_df3.columns=common_colname
delete_paran(new_df3)
```

```python
def delete_wave(df) : # 여러 값이 있는 것들(10~15) 지우기
    for i in range(df["sugar"].shape[0]):
        for string in ("sugar","calories"):
            data=df[string][i]
            index=data.find("~")
            if index != -1 :
                df[string][i]=data[:index]
delete_wave(new_df3)
```

```python
df4["restaurant"]="macdonald"
# 이상한 공백 값이 있어서, 직접 조회하여 열 순서 바꿈
new_df4=df4[[df4.columns[-1],"Item","Calories","Sodium","Sugars","Total Fat","Protein"]]
new_df4.columns=common_colname
```

```python
df5["restaurant"]="no_brand_burger"
# 이상한 공백 값이 있어서, 직접 조회하여 열 순서 바꿈
new_df5=df5[[df5.columns[-1],df5.columns[0],df5.columns[2],df5.columns[-2],df5.columns[3],df5.columns[5],df5.columns[4]]]
new_df5.columns=common_colname

for i in range(new_df5["sodium"].shape[0]):
    new_df5["sodium"].iloc[i]=new_df5["sodium"].iloc[i].replace(",","")
for i in range(new_df5["calories"].shape[0]):
    new_df5["calories"].iloc[i]=new_df5["calories"].iloc[i].replace(",","")
```

```python
df6["restaurant"]="prank_burger"
# 이상한 공백 값이 있어서, 직접 조회하여 열 순서 바꿈
new_df6=df6[[df5.columns[-1],df5.columns[0],df5.columns[2],df5.columns[3],df5.columns[4],df5.columns[5],df5.columns[-2]]]
new_df6.columns=common_colname
```

```python
df=pd.concat((df1,new_df2,new_df3,new_df4,new_df5,new_df6))
```

```python
df=df[df["protein"].isna() == False].reset_index(drop=True) # 결측값 제거
df.to_csv("Hamburger.csv", index=False)
```
