```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

html=urlopen("https://www.lotteeatz.com/upload/stg/etc/ria/items.html")
soup=BeautifulSoup(html, "html.parser")
title = soup.select("body > table > thead > tr > th")
columns=[]
for i in title:
    columns.append(i.string)
```

```python
content=soup.select("tbody > tr > td")
content2 = [td for td in content if 'rowspan' not in td.attrs]
content=[]
text=[]
for i in content2:
    content.append(i.string)
    if len(content)==10:
        text.append(content)
        content=[]
```

```python
import pandas as pd
lotteria=pd.DataFrame(text,columns=columns[1:])
lotteria.to_csv("../Data/lotteria.csv",index=False,encoding='utf-8-sig')
```

```python
html=urlopen("https://emilysinglelife.tistory.com/62")
soup=BeautifulSoup(html, "html.parser")

find_text='''#article-view > div.tt_article_useless_p_margin.contents_style > table:nth-child(27),
#article-view > div.tt_article_useless_p_margin.contents_style > table:nth-child(39),
#article-view > div.tt_article_useless_p_margin.contents_style > table:nth-child(52),
#article-view > div.tt_article_useless_p_margin.contents_style > table:nth-child(63)'''

title = soup.select(find_text)
# table:nth-child에 들어가는 숫자가 27, 39, 52, 63인 요소만 뽑아내기
content=[]
text=[]
for s in title:
    for i in s:
        for j in i.select("td") :
            # trim을 주어서 양쪽 공백을 제거
            msg=j.text.strip()

            if msg not in ["카페인(mg)",""]:
                content.append(msg)
            if len(content)==7:
                text.append(content)
                content=[]
```

```python
burgerking=pd.DataFrame(text[1:],columns=text[0]).drop([87,88,89])
```

```python
burgerking[burgerking["제품명"]!="제품명"].reset_index(drop=True).to_csv("../Data/burgerking.csv",index=False,encoding='utf-8-sig') 
```

```python
import pytesseract
from PIL import Image, ImageFilter

pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
image_path = 'moms_touch.png'
image = Image.open(image_path)

# 이미지의 크기를 확장
new_width = image.width * 3
new_height = image.height * 3
image = image.resize((new_width, new_height), Image.BICUBIC)

# OCR 수행 (lang 옵션에 'kor' 추가)
text = pytesseract.image_to_string(image, lang='kor')
with open("moms_touch.txt", "w") as f: # 파일에 쓰기 
    f.write(text)
```
